Question 5:

```python
import numpy as np
from scipy.stats import binom

# Define the parameters
n = 124  # Number of trials (couples)
p = 0.5  # Probability of success (tilting right under H0)
observed_tilts_right = 80  # Observed number of couples tilting to the right

# Calculate the p-value using the binomial distribution's survival function (1 - CDF)
# This gives the probability of getting 80 or more tilts to the right
p_value = binom.sf(observed_tilts_right - 1, n, p)

# Output the result
print(f"The p-value for observing 80 or more right-tilting couples is: {p_value}")
```

^ Showing there is very strong evidence aginst the null hypothesis

Question 8:

```python
import numpy as np
from scipy.stats import binom

# Parameters
n = 80  # number of students
p_null = 0.5  # probability under the null hypothesis
observed_successes = 49  # observed correct identifications

# Calculate p-value (P(X >= observed_successes | H0))
p_value = 1 - binom.cdf(observed_successes - 1, n, p_null)

# Output results
print(f"P-value: {p_value:.4f}")
```

```python
import matplotlib.pyplot as plt

# Binomial distribution under H0
x = np.arange(0, 81)
y = binom.pmf(x, n, p_null)

plt.bar(x, y, color='lightblue', edgecolor='black')
plt.axvline(observed_successes, color='red', linestyle='--', label='Observed 49 successes')
plt.xlabel('Number of Correct Identifications')
plt.ylabel('Probability')
plt.title('Binomial Distribution Under Null Hypothesis (p=0.5)')
plt.legend()
plt.show()
```
